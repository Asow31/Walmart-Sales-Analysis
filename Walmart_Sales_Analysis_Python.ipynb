```python
import pandas as pd

df = pd.read_csv("Walmart_Sales_Small.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
store_sales = df.groupby("Store")["Weekly_Sales"].sum().sort_values(ascending=False)
store_sales
```

```python
dept_sales = df.groupby("Dept")["Weekly_Sales"].sum().sort_values(ascending=False)
dept_sales
```

```python
holiday_sales = df.groupby("IsHoliday")["Weekly_Sales"].sum()
holiday_sales
```

```python
import matplotlib.pyplot as plt

store_sales.plot(kind="bar", figsize=(12,5))
plt.title("Total Weekly Sales by Store")
plt.xlabel("Store")
plt.ylabel("Total Sales")
plt.show()
```

```python
dept_sales.head(10).plot(kind="bar", figsize=(12,5))
plt.title("Top 10 Departments by Total Sales")
plt.xlabel("Department")
plt.ylabel("Total Sales")
plt.show()
```
